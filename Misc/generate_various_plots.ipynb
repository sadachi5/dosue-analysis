# Sine wave

```python
def sin(amp, t, freq, t0):
    return amp * np.sin( (2.*np.pi)*freq * (t - t0) )
```

# Planck's law

```python
## planckEquation() ##
# nu : frequency [Hz=/sec]
# stefanBoltzmannEquation = pi*Integral(0,+inf)(planck)dnu
def planckEquation(nu, T=300):
    h = 6.626e-34 ; # [Jsec] Planck constant
    c = 3.0e+8    ; # [m/sec] light velocity 
    k = 1.380e-23 ; # [J/K] Boltzmann constant
    hvkT = h*nu/(k*T); # [J/J]
    I = 2.*h*(nu**3.)/(c**2.) * 1./(np.exp(hvkT)-1.) ; # [Jsec /sec /sec /sec     sec/m sec/m=J/m^2=(J/sec)*(sec)*(/m^2)=W/(Hz*m^2)]
    return I;
## end of planckEquation() ##
```

```python
freq = np.linspace(0.,100e+9,10001)
```

```python
plt.plot(freq, planckEquation(freq, 10))
plt.plot(freq, planckEquation(freq, 20))
plt.plot(freq, planckEquation(freq, 40))
plt.plot(freq, planckEquation(freq, 80))
plt.plot(freq, planckEquation(freq, 160))
plt.plot(freq, planckEquation(freq, 320))
plt.xscale('log')
plt.yscale('log')
```
