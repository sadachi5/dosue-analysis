# Downconvertion (heterodyne)

```python
import os
import sys
import numpy as np
from matplotlib import pyplot as plt
```

```python
time = np.linspace(0,0.5,1000)
freq_rf = 100
freq_lo = 90
A_rf = 1
A_lo = 1
rf = A_rf * np.sin(2*np.pi*time*freq_rf)
lo = A_lo * np.sin(2*np.pi*time*freq_lo)
mix = rf + lo
print(time)
print(rf)
print(lo)
```

```python
fig = plt.figure(figsize=(10,4))
plt.plot(time, rf, c='tab:red')
plt.plot(time, lo, c='tab:pink')
plt.plot(time, mix, c='tab:blue')
plt.plot([time[0],time[-1]], [0,0], c='black', lw=0.5)
plt.xlim(time[0],time[-1])
```

```python
fig = plt.figure(figsize=(10,4))
plt.plot(time, rf, c='tab:red', lw=2)
plt.plot(time, lo, c='tab:pink', lw=2)
plt.plot([time[0],time[-1]], [0,0], c='black', lw=0.5)
plt.xlim(time[0],time[-1])
```
